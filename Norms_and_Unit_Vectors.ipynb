```python
import numpy as np # import numpy dari library
x = np.array([5, 55, 61]) #inisialisasi vektor
print(x) #tampilkan vektor
print(np.linalg.norm(x)) #hitung l2 norm

v = np.array([3, 4]) #inisialisasi vektor v
norm = np.linalg.norm(v)
print(np.linalg.norm(v)) #hitung l2 norm
unit_v = v / norm #buat unit vektor dan masukkan ke unit_v
print(unit_v) #tampilkan unit vektor
print(np.linalg.norm(unit_v)) #tampilkan panjang unit vektor
```
