```python
import numpy as np
i = np.array([1, 0]) #inisialisasi vektor (1,0)
print(i) #tampilkan i
j = np.array([0, 1]) #inisialisasi vektor (0,1)
print(j) #tampilkan j
np.dot(i, j) #mebuktikan vektor i dan j ortogonal

v = np.array([1, 2]) #inisialisasi vektor (1,2)
print(v) #tampilkan v
w = np.array([2, -1]) #inisialisasi vektor (2,-1)
print(w) #tampilkan w
np.dot(v, w) #mebuktikan vektor v dan w ortogonal
```
